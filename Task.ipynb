```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

titanic_subset = titanic[["age", "sex"]]
titanic_clean = titanic_subset.dropna(subset=["age"])

titanic_clean.hist(column="age", by="sex", bins=50, figsize=(30,15), edgecolor="black")
titanic_subset = titanic[["age", "sex"]]
print(titanic_subset.head())

plt.suptitle("Age Distribution of Titanic Passengers by Sex")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```
